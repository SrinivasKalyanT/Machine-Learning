```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("F:/Datasets/heart_disease_uci.csv")
```

```python
df
```

```python
df = df.drop(columns = ['id'])
```

```python
df
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['sex'] = df['sex'].astype('category')
df['sex'] = df['sex'].cat.codes
df
```

```python
df['dataset'] = df['dataset'].astype('category')
df['dataset'] = df['dataset'].cat.codes
df
```

```python
df['cp'] = df['cp'].astype('category')
df['cp'] = df['cp'].cat.codes
df
```

```python
df['fbs'] = df['fbs'].astype('category')
df['fbs'] = df['fbs'].cat.codes
df
```

```python
df['thal'] = df['thal'].astype('category')
df['thal'] = df['thal'].cat.codes
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df = df[df['thal']!=-1]
```

```python
x = df.drop(columns='thal')
x
```

```python
y = df['thal']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
```

```python
x_train = Scaler.fit_transform(x_train)
```

```python
x_test = Scaler.transform(x_test)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameter = {'penalty' : ['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,10,100],'max_iter':[100,200,300]}
```

```python
classifier_regressor = GridSearchCV(classifier,param_grid=parameter,scoring = 'accuracy',cv = 5 )
```

```python
classifier_regressor.fit(x_train,y_train)
```

```python
print(classifier_regressor.best_params_)
```

```python
print(classifier_regressor.best_score_)
```

```python
y_pred = classifier_regressor.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_pred,y_test)
score
```

```python
cls = classification_report(y_pred,y_test)
cls
```
